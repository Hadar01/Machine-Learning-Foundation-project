#Here is a function to check the data of each country

```python
import matplotlib.pyplot as plt

def plot_meat_consumption_by_country(country_name, data):
   
    # Filter the data for the specified country
    country_data = data[data['Entity'] == country_name]

    if country_data.empty:# Check if the country data is empty
        print(f"No data available for '{country_name}'. Please check the country name.")
        return

    # List of meat types to plot
    meat_types = ['Poultry', 'Beef', 'Sheep and goat', 'Pork', 'Other meats', 'Fish and seafood']

    # Create subplots
    plt.figure(figsize=(14, 10))

    for i, meat in enumerate(meat_types, 1):
        plt.subplot(3, 2, i)  # Create a grid of 3 rows and 2 columns
        plt.plot(country_data['Year'], country_data[meat], marker='o', label=meat)
        plt.title(f'{meat} Consumption in {country_name}')
        plt.xlabel('Year')
        plt.ylabel('Consumption')
        plt.grid(True)
        plt.tight_layout()

    # Display the plots
    plt.suptitle(f'Meat Consumption Trends in {country_name}', fontsize=16, y=1.02)
    plt.show()
```

```python
plot_meat_consumption_by_country('Afghanistan', df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def compare_meat_consumption(countries, data):
   
    # Filter the data for the selected countries
    comparison_data = data[data['Entity'].isin(countries)]

    if comparison_data.empty:
        print("No data available for the specified countries. Please check the country names.")
        return

    # List of meat types to compare
    meat_types = ['Poultry', 'Beef', 'Sheep and goat', 'Pork', 'Other meats', 'Fish and seafood']

    # Set up the overall figure
    plt.figure(figsize=(16, 12))

    # Plot trends for each meat type
    for i, meat in enumerate(meat_types, 1):
        plt.subplot(3, 2, i)  # Create a grid of 3 rows and 2 columns
        sns.lineplot(data=comparison_data, x='Year', y=meat, hue='Entity', marker='o')
        plt.title(f'{meat} Consumption Trends')
        plt.xlabel('Year')
        plt.ylabel('Consumption')
        plt.legend(title='Country')
        plt.grid(True)

    # Adjust layout and display the plots
    plt.suptitle('Meat Consumption Trends Comparison', fontsize=18, y=1.02)
    plt.tight_layout()
    plt.show()
```

```python
compare_meat_consumption(['Afghanistan', 'United States', 'India'], df)
```
